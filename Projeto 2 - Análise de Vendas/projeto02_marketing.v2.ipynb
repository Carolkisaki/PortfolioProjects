## Importando bibliotecas

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

## Importando a Base de Dados

```python
df = pd.read_csv('C:\PYTHON\projeto1\dados_marketing.csv', sep=';')
display(df)
```

## Tratamento de Dados
### Observando a tabela, a coluna Comprou mostra apenas os números 0 e 1, para facilitar na hora de criar os graficos, faço a alteração de 0 para Não e 1 para sim

```python
df['Comprou'] = df['Comprou'].replace({0:'Não', 1:'Sim'})
display(df)
```

### Observa-se que no banco de dados há informações sobre gastos com alimentos, brinquedos, eletrônicos, móveis, utilidades e vestuários, para resumir os dados e então cruzar esses dados com outras variáveis criarei uma coluna chamada TotalGasto

```python
df['TotalGasto'] =(
    df['Gasto com Alimentos'] +
    df['Gasto com Brinquedos'] +
    df['Gasto com Eletronicos'] +
    df['Gasto com Moveis'] +
    df['Gasto com Utilidades'] +
    df['Gasto com Vestuario']
)
display(df.head())
```

### Percebe-se também que a coluna 'Data Cadastro', manteve o dia, mês e ano na mesma coluna, pensando em analisar os dados, afim de comparar por ano, crio uma nova coluna apenas com o ano

```python
df['Data Cadastro'] = pd.to_datetime(df['Data Cadastro'], dayfirst = True)
df['Ano'] = df['Data Cadastro'].dt.year
display(df.head())
```

## Iniciando a construção dos gráficos
### Mas por que usar o Seaborn, ao invés do Plotly? 
### Apesar do Plotly gerar gráficos interativos e ser fácil de usar, infelizmente, o Github não aceita seus gráficos na hora de importar um projeto para o repositório. O Matplotlib é a biblioteca mais comum, sendo a base da visualização de dados, inclusive o Seaborn e o plot do Pandas são produzidos em cima da biblioteca. Ideal para visualizações rápidas, para poder melhorar o design dos gráficos precisa fazer a customização manual, detalhe que outras bibliotecas já fazem automaticamente. O Seaborn já faz gráficos mais bonitos esteticamente e é bem integrado com o Pandas

```python
dados_agrupados = df.groupby('Escolaridade', as_index=False)['ID'].sum()
fig = sns.barplot(data = dados_agrupados, x = 'Escolaridade', y = 'ID', ci = None, palette = 'viridis')
plt.title('Clientes x Escolaridade', fontsize = 16)
plt.show()
```

```python
dados_agrupados = df.groupby('Estado Civil', as_index = False)['ID'].sum()
fig = sns.barplot(data = dados_agrupados, x = 'Estado Civil', y = 'ID', palette = 'Blues')
plt.title('Clientes x Estado Civil', fontsize = 16)
plt.show()
```

```python
totalgasto_agrupado = df.groupby('Filhos em Casa', as_index = False)['TotalGasto'].sum()
fig = sns.barplot(data = totalgasto_agrupado, x = 'Filhos em Casa', y = 'TotalGasto')
plt.title('Total de Gasto x Filhos em Casa', fontsize = 16)
plt.show()
```

```python
totalgasto_agrupado = df.groupby('Adolescentes em Casa', as_index = False)['TotalGasto'].sum()
fig = sns.barplot(data = totalgasto_agrupado, x = 'Adolescentes em Casa', y = 'TotalGasto')
plt.title('Total de Gasto x Adolescentes em Casa', fontsize = 16)
plt.show()
```

### Será que quanto maior o salário anual, maior será o Total Gasto? Para isso utilizarei o gráfico de dispersão que é interessante quando queremos ver a relação entre duas ou mais variáveis.

```python
fig = sns.scatterplot(data=df, x='Salario Anual', y='TotalGasto')
plt.title('Total de Gasto x Salário Anual', fontsize = 16)
plt.show()
```

### Observe que há um valor atípico no gráfico de dispersão, o famoso outlier, então definirei um limite para o salário anual, para que então possa excluir esse valor anômalo¶

```python
limite_salario = 163000

data_filtrada = df[df['Salario Anual'] <= limite_salario]
```

```python
fig = sns.scatterplot(data=data_filtrada, x='Salario Anual', y='TotalGasto')
plt.title('Total de Gasto x Salário Anual', fontsize = 16)
plt.show()
```

### Agora vamos ver como foi a performance das campanhas
### Efetividade da Campanha x Número de Filhos = Será que as pessoas que possuem filhos comprou mais do que os que não possuem?

```python
fig = sns.catplot(data = df, kind = 'bar', x = 'Filhos em Casa', y = 'TotalGasto', hue = 'Comprou', palette = 'flare')
plt.title('Efetividade da Campanha x Número de Filhos')
plt.show()
```

### Será que quanto maior o salário anual, maiores são as chances de compras?

```python
media_salarioanual = df.groupby('Comprou')['Salario Anual'].mean().reset_index()
```

```python
fig = sns.barplot(data = media_salarioanual, x = 'Comprou', y = 'Salario Anual')
plt.title = "Média de Salário Anual x Resultado das Campanhas"
plt.show()
```

```python
fig = sns.relplot(data = df, kind = 'line', x = 'Ano', y = 'TotalGasto', hue = 'Pais', linewidth=2.5)
plt.title = 'Total de Gasto por Ano e País'
plt.show()
```
